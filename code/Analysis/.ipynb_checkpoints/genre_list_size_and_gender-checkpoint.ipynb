This notebook examines how genre list length correlates with gender.

The list length counts by gender find that in the 11+ range the expected of male to female is 1. However, there are no female artists with > 11 labels. The statistics don't capture the fact that male artists have features that are not available to female artists. By switching to 12+ the inability to carry out the calculation (divide by zero) is informative. 

The data for lists of length > 6 needs to be dealt with carefully. It is exactly at that point that the statistical significant comes into doubt. First, using 10-fold CV to estimate the ratios there is often a zero value which makes the ratio undefined. Second, when filtering the data to only include, e.g. lengths > 5, then the mean of the error and std of the error indicate that there is not a clear effect.

### Genre Labels

Each value of the genre column is a _string_ of comma separated genre labels. We want to convert it to a _list_ of strings.

```python
"""This function takes in a string of the form
appearing in the genrelist of the dataframe.
It strips the square brackets and extra quotes and
returns a list of strings where each string is a genre label."""
def genrelist(string):
    string = string.strip("[").strip("]").replace("'","")
    L = [s for s in string.split(',')]
    L_new = []
    for x in L:
        L_new.append(x.replace(" ","_").lstrip("_").rstrip("_"))
    while (str("") in L_new):
        L_new.remove("")
    return L_new
```

Now we apply it to the whole column and put the lists in a new column:

```python
data['genrelist']= data['genrelist'].apply(genrelist)
```

### This function creates a DF with the frequencies of the lengths of lists by gender for a df of the type of "data" above:

```python
def create_length_counts_by_gender(df):
    """df is """
    
    #length counts total
    dft = df.groupby(['genrelist_length']).count()
    dft.drop(['gender'], axis = 1, inplace = True)
    dft.columns = ['artist_count']
    
    # length counts by gender
    df = df.groupby(['genrelist_length','gender']).count()
    df.columns = ['artist_count']

    df.reset_index(inplace = True)
    df.set_index(['genrelist_length'], inplace = True)
    df = df.pivot(columns = 'gender')
    
    #flatten index and replace with single strings
    df.columns = df.columns.to_flat_index()
    df.columns = [name[1] for name in df.columns]
    #combine total count with gender counts
    dft.columns = ['total']
    df = dft.join(df, how = 'outer')
    
    # bin 11 count and above
    df.at['11+'] = df.iloc[10:,:].sum()
    
    # calculated columns: relative and ratios
    df['female_relative'] = round(df['female']/fem,3)
    df['male_relative'] = round(df['male']/mal,3)

    df['ratio'] = round(df['male_relative']/df['female_relative'],3)

    df.fillna(value = {'female': 0}, inplace = True)
    
    return df
```

Estimate ratio using cross validation:

```python
ratios = pd.DataFrame({'ratio':lcbg['ratio']})
ratios_test = pd.DataFrame({'ratio':lcbg['ratio']})

kf = KFold(n_splits = 10, shuffle = True, random_state = seed)

fold = 0
for train, test in kf.split(data.genrelist_length.values): #use any column to define the train, test index sets
    data_train = data.iloc[train]
    data_test = data.iloc[test]
    
    #train length counts
    length_counts = create_length_counts_by_gender(data_train)
    ratios = ratios.join(length_counts['ratio'], how = 'outer', rsuffix = fold+1)
    
    #test length counts
    length_counts = create_length_counts_by_gender(data_test)
    ratios_test = ratios_test.join(length_counts['ratio'], how = 'outer', rsuffix = fold+1)
    fold += 1
    
ratios.drop(['ratio'], axis = 1, inplace = True)
ratios_test.drop(['ratio'], axis = 1, inplace = True)

# calculate mean and std of CV errors
errors = ratios-ratios_test
# mean
errors_mean = errors.T.mean().to_frame()
errors_mean.columns = ['mean']
errors = errors.join(errors_mean, how = 'outer')

# std
errors_std = errors.T.std().to_frame()
errors_std.columns = ['std']
errors = errors.join(errors_std, how = 'outer')

# calculate mean of counts for each list length
df_mean = ratios.T.mean().to_frame()
df_mean.columns = ['mean']
ratios = ratios.join(df_mean, how = 'outer')

# calculate std of counts for each list length
df_std = ratios.T.std().to_frame()
df_std.columns = ['std']
ratios = ratios.join(df_std, how = 'outer')

ratio_summary = ratios[['mean','std']].join(errors[['mean','std']], how = 'outer', rsuffix = '_error')
```

```python
def cv_estimates(df1,df2):
    ratios = pd.DataFrame({'ratio':df1['ratio']})
    ratios_test = pd.DataFrame({'ratio':df1['ratio']})

    kf = KFold(n_splits = 5, shuffle = True, random_state = seed)

    fold = 0
    for train, test in kf.split(df2.genrelist_length.values): #use any column to define the train, test index sets
        data_train = df2.iloc[train]
        data_test = df2.iloc[test]

        #train length counts
        length_counts = create_length_counts_by_gender(data_train)
        ratios = ratios.join(length_counts['ratio'], how = 'outer', rsuffix = fold+1)

        #test length counts
        length_counts = create_length_counts_by_gender(data_test)
        ratios_test = ratios_test.join(length_counts['ratio'], how = 'outer', rsuffix = fold+1)
        fold += 1

    ratios.drop(['ratio'], axis = 1, inplace = True)
    ratios_test.drop(['ratio'], axis = 1, inplace = True)

    # calculate mean and std of CV errors
    errors = ratios-ratios_test
    # mean
    errors_mean = errors.T.mean().to_frame()
    errors_mean.columns = ['mean']
    errors = errors.join(errors_mean, how = 'outer')

    # std
    errors_std = errors.T.std().to_frame()
    errors_std.columns = ['std']
    errors = errors.join(errors_std, how = 'outer')

    # calculate mean of counts for each list length
    df_mean = ratios.T.mean().to_frame()
    df_mean.columns = ['mean']
    ratios = ratios.join(df_mean, how = 'outer')

    # calculate std of counts for each list length
    df_std = ratios.T.std().to_frame()
    df_std.columns = ['std']
    ratios = ratios.join(df_std, how = 'outer')

    ratio_summary = ratios[['mean','std']].join(errors[['mean','std']], how = 'outer', rsuffix = '_error')
    return ratios, errors, ratio_summary
```

```python
ratios, errors, ratio_summary = cv_estimates(lcbg, data_long)
```
